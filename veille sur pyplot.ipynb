### Pyplot est une sous-librairie de matplotlib qui elle, est une librairie de python qui permet d'effectuer des visualisations graphiques. Dans les lignes de code qui suivent, nous donnons un exemple de visualisation graphique à partir de pyplot.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = [150, 120]
labels = ['absence','presence']
colors = ['tab:blue', 'tab:cyan']

fig, ax = plt.subplots()
ax.pie(x, labels = labels, colors = colors)
ax.set_title('repartition presence et absence de maladie')
plt.show()
```
